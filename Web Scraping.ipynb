# Importing libraries and modules

```python
#Importing

import pandas as pd
import requests
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

# Initializing web scraping

```python
#Saving the url we want to scrap

url='your_Airbnb_url'

#Requesting website

req=requests.get(url)

html=req.content



#HTML parsing

page_soup=soup(html,'html.parser')

#Capturing every container of the Stay

containers=page_soup.findAll('div',{'class':'_8ssblpx'})
```

# Gathering Data

```python
#Initializing null lists to save info that is captured from the website

first_description=[]
main_description=[]
features=[]
amenities=[]
stars=[]
reviews=[]
price=[]


#This is the piece of code that capture the data

#While loop that iterates over all listings in the search: Eg: NYC has 20 Stays per sheet, there are almost 15 pages for total
# of 300 results

while True:
    
    
#For loop that extracts infomartion in every sheet

    for container in containers:

        first_description_container=container.findAll('div',{'class':'_b14dlit'})
        first_description.append(first_description_container[0].text)

        main_description_container=container.findAll('div',{'class':'_bzh5lkq'})
        main_description.append(main_description_container[0].text)

        features_container=container.findAll('div',{'class':'_kqh46o'})
        features.append(features_container[0].text)

        amenities_container=container.findAll('div',{'class':'_kqh46o'})
        amenities.append(amenities_container[1].text)

        stars_container=container.findAll('span',{'class':'_10fy1f8'})

        if stars_container==[]:

            stars.append('no_stars')

        else:
            stars.append(stars_container[0].text)


        reviews_container=container.findAll('span',{'class':'_a7a5sx'})

        if reviews_container==[]:

            reviews.append('no_reviews')

        else:

            reviews.append(reviews_container[0].text.strip('\xa0,(,)'))

        price_container=container.findAll('span',{'class':'_1p7iugi'})
        price.append(price_container[0].text.strip(':'))
    
    
    #Back in the while loop: the try statement evaluate if an error occurs:
    
    try:
        
        #Extract the pice of url that take us to the next sheet
        next_page=page_soup.findAll('div',{'class':'_jro6t0'})[0].find('a',{'class':'_za9j7e'})['href']
    
    
    #The except clause allow us to do actions if an error occurs: in this case activates the break stamentent which allows us
    # to interupt the while because there are no more sheets to search Stays
    except:
        
        break
    
    #Getting the next sheet url
    
    next_url='https://www.airbnb.com'+next_page
    
    #Repeat the steps in the above cell
    
    req=requests.get(next_url)

    html=req.content

    #HTML parsing

    page_soup=soup(html,'html.parser')
    
    containers=page_soup.findAll('div',{'class':'_8ssblpx'})
    
    
    
    
 
#Building final dataframe


data=pd.DataFrame({'first_description':first_description,'main_description':main_description,'features':features,
                    'amenities':amenities,'stars':stars,'reviews':reviews,'price':price
                   })


                                            
```
